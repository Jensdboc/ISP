# Capture video

```python
# Initialize the camera
cap = cv2.VideoCapture(0)  # 0 indicates the default camera (you can change this number if you have multiple cameras)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Process the frame (e.g., resize, normalize, etc.) as needed
    processed_frame = cv2.resize(frame, (256, 256))
    processed_frame = processed_frame / 255.0
    processed_frame = np.expand_dims(processed_frame, axis=0)
    predictions = model.predict(processed_frame)
    c = np.argmax(predictions, axis=1)
    text = classes[int(c)]
    print(text)
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 0.75
    font_thickness = 2
    color = (128, 0, 128)
    position = (10, 30)  # Coordinates (x, y) of the upper left corner
    
    # Add text to the image
    cv2.putText(frame, text, position, font, font_scale, color, font_thickness)
    
    cv2.imshow('Camera Feed', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close the window
cap.release()
cv2.destroyAllWindows()
```

# Evaluate test set

```python
test_path = "D:\Documents\GitHub\ISP\\test"

for dirpath, dirnames, filenames in os.walk(test_path):
    for filename in filenames:
        test_image_path = test_path + "\\" + filename
        
        image = tf.io.read_file(test_image_path)
        image = tf.image.decode_image(image, channels=3)  # Channels=3 for RGB images
        image = tf.image.resize(image, [256, 256])  # Resize to match model input size
        image = image / 255.0  # Normalize
        image = tf.expand_dims(image, axis=0)  # Add batch dimension if needed
        
        predictions = model.predict(image)
        c = np.argmax(predictions, axis=1)
        display(Image(test_image_path))
        # print(predictions)
        # print(c)
        print(f"{filename[:-4]} -> {classes[int(c)]}")
```

# Evaluate test set

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import scipy

test_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_generator = test_datagen.flow_from_directory(
    "D:\Documents\\Uniefwerk\\5de_jaar\\1ste_semester\ISP\Project\Data\Dataset1\\New Plant Diseases Dataset(Augmented)\\New Plant Diseases Dataset(Augmented)\\valid",
    target_size=(256, 256),
    batch_size=32,
    class_mode='sparse'
)

result = model.evaluate(test_generator)

print("Test Loss:", result[0])
print("Test Accuracy:", result[1])
```

# Extra

```python
predicted_labels = model.predict(test_generator)
true_labels = test_generator.classes

# Get the class with the highest probability as the predicted class
predicted_labels = np.argmax(predicted_labels, axis=1)

print(predicted_labels)
print(true_labels)
```

```python
# Construct confusion matrix
confusion_mat = np.zeros((38, 38))

for i in range(len(true_labels)):
    confusion_mat[true_labels[i]][predicted_labels[i]] += 1

print("Confusion Matrix:")
print(confusion_mat)
```

```python
for i in range(38):
    print(confusion_mat[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the heatmap
im = ax.imshow(confusion_mat, cmap='Blues')

# Add colorbar
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel("Counts", rotation=-90, va="bottom")

# Set axis labels
ax.set_xlabel("Predicted")
ax.set_ylabel("True")

# Set class labels on axis ticks (optional)
if classes is not None:
    ax.set_xticks(np.arange(len(classes)))
    ax.set_yticks(np.arange(len(classes)))
    ax.set_xticklabels(classes)
    ax.set_yticklabels(classes)

# Loop over data dimensions and create text annotations
for i in range(len(classes)):
    for j in range(len(classes)):
        text = ax.text(j, i, confusion_mat[i, j],
                       ha="center", va="center", color="w")

# Display the heatmap
plt.show()
```
